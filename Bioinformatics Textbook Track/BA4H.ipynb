# BA4H
Generate the Convolution of a Spectrum

```python
def spec_convolution(spectrum):
    out = []
    spectrum.sort()
    pre = spectrum[:-1]
    suf = spectrum[1:]
    table = {}
    for i in range(len(pre)):
        for j in range(i, -1, -1):
            if suf[i] > pre[j]:
                mass = suf[i] - pre[j]
                out.append(mass)
                if mass not in table.keys():
                    table[mass] = 0
                table[mass] += 1
    res = []
    allkeys = sorted(table, key = table.get, reverse = True)
    for k in allkeys:
        res.extend([k] * table[k])
    return res
```

```python
with open('rosalind_ba4h.txt', 'r') as f:
    spectrum = f.readline().rstrip().split()
    spectrum = [int(k) for k in spectrum]
    out = spec_convolution(spectrum)
    print(' '.join([str(i) for i in out]))
```

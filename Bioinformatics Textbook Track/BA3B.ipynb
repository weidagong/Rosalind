# BA3B Reconstruct a String from its Genome Path

```python
def BA3B(filename):
    with open(filename, 'r') as f:
        dna = []
        for line in f:
            dna.append(line.rstrip())
        genome = dna[0]
        for i in range(1, len(dna), 1):
            genome += dna[i][-1]
        return(genome)
```

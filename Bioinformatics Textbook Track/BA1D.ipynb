# BA1D Find All Occurrences of a Pattern in a String

```python
def BA1D(filename):
    with open(filename, 'r') as f:
        pattern = f.readline().rstrip()
        genome = f.readline().rstrip()
        start = 0
        pos = []
        while True:
            start = genome.find(pattern, start) + 1
            if start == 0:
                break
            else:
                pos.append(start - 1)
        pos = ' '.join([str(x) for x in pos])
        return(pos)
        
```

# BA5F Local Alignment

```python
import pandas as pd


scoreMatrix62 = pd.read_csv('scoreMatrix62.txt', delim_whitespace=True)
scoreMatrix250 = pd.read_csv('scoreMatrix250.txt', delim_whitespace=True)
```

```python
import numpy as np


def LocalAlignment(v, w, scoreMatrix):
    score = np.zeros([len(v) + 1, len(w) + 1])
    backtrack = [[] for _ in range(len(v) + 1)]
    backtrack[0].append('-')
    for i in range(1, len(v) + 1):
        score[i, 0] = 0
        backtrack[i].append('-')
    for j in range(1, len(w) + 1):
        score[0, j] = 0
        backtrack[0].append('-')
    for i in range(1, len(v) + 1):
        for j in range(1, len(w) + 1):
            diag = score[i - 1, j - 1] + scoreMatrix[v[i - 1]][w[j - 1]]
            horizontal = score[i, j - 1] - 5
            vertical = score[i - 1, j] - 5
            score[i, j] = max(diag, horizontal, vertical, 0)
            if score[i, j] == score[i - 1,j] - 5:
                backtrack[i].append("↓")
            elif score[i, j] == score[i, j - 1] - 5:
                backtrack[i].append("→")
            elif score[i, j] == score[i - 1, j - 1] + scoreMatrix[v[i - 1]][w[j - 1]]:
                backtrack[i].append("↘")
            elif score[i, j] == 0:
                backtrack[i].append('-')
    maxScore = np.amax(score)
    coordinates = np.where(score == maxScore)
    x = coordinates[0][0]
    y = coordinates[1][0]
    return score, maxScore, backtrack, x, y        
```

```python
def IterOutputLCSLocal(backtrack, v, w, i, j, score):
    first = ''
    second = ''
    while i > 0 or j > 0:
        if backtrack[i][j] == "-" or score[i, j] == 0:
            break
        elif backtrack[i][j] == "↓":
            first += v[i -1]
            second += '-'
            i = i - 1
        elif backtrack[i][j] == "→":
            first += '-'
            second += w[j - 1]
            j = j - 1
        elif backtrack[i][j] == "↘":
            first += v[i - 1]
            second += w[j - 1]
            i = i - 1
            j = j - 1
        
    first = first[::-1]
    second = second[::-1]
    return first, second
```

```python
with open('rosalind_ba5f.txt', 'r') as f:
    v = f.readline().rstrip()
    w = f.readline().rstrip()
    score, maxScore, backtrack,x,y = LocalAlignment(v, w, scoreMatrix250)
    print(maxScore)
    first, second = IterOutputLCSLocal(backtrack, v, w, x, y, score)
    print(first)
    print(second)
```

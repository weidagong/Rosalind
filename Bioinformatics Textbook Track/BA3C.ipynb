# BA3C Construct the Overlap Graph of a Collection of k-mers

```python
def overlap(a, b):
    if a[1:] == b[:len(b) - 1]:
        return(True)
    return(False)

def BA3C(filename):
    with open(filename, 'r') as f:
        kmers = [line.rstrip() for line in f]
        for i in range(len(kmers)):
            for j in range(len(kmers)):
                if i == j:
                    continue
                if overlap(kmers[i], kmers[j]):
                    print('%s -> %s ' % (kmers[i], kmers[j]))
```

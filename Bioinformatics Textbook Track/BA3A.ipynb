# BA3A Generate the k-mer Composition of a String

```python
def BA3A(filename):
    with open(filename, 'r') as f:
        k = int(f.readline().rstrip())
        seq = f.readline().rstrip()
        kmers = []
        for i in range(len(seq) - k + 1):
            kmers.append(seq[i : i + k])
        kmers.sort()
        for kmer in kmers:
            print(kmer)
```

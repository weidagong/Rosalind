# BA5E Global Alignment

```python
from Bio import pairwise2
from Bio.SubsMat import MatrixInfo as matlist

v = 'PLEASANTLY'
w = 'MEANLY'
for a in pairwise2.align.globalds(v, w, matlist.blosum62, -5, -5):
    print("{}\n{}\n{}\n\n".format(int(a[2]), a[0][a[3]:a[4]], a[1][a[3]:a[4]]))
    break  # only need first one
```

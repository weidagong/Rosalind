# BA3D Construct the De Bruijn Graph of a String

```python
from collections import defaultdict

def deBrujin(k, genome):
    kmers = defaultdict(list)
    for i in range(len(genome) - (k - 1)):
        kmers[genome[i : i + k - 1]].append(genome[i + 1: i + k])
    return(kmers)

def BA3D(filename):
    with open(filename, 'r') as f:
        k = int(f.readline().rstrip())
        genome = f.readline().rstrip()
        kmers = deBrujin(k, genome)
        for k, v in kmers.items():
            out = ','.join(v)
            print('%s -> %s' % (k, out))
            
            
```
